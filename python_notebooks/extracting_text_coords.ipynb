```python
import fitz
import cv2
import matplotlib.pyplot as plt
import matplotlib.patches as patches

sample_pdf = r"C:\Users\Conno\OneDrive\Desktop\DE Textbook.pdf"
pdf = fitz.open(sample_pdf)
page = pdf[0]

#print the page
#img = cv2.imread('output_page1.png')
#img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
#fig, ax = plt.subplots(figsize=(12, 12))
#ax.imshow(img_rgb)


words = page.get_text("words") # Returns list of tuples: (x0, y0, x1, y1, word, block_no, line_no, word_no)
print(len(words))
for x0, y0, x1, y1, word, block_no, line_no, word_no in words:
    print(f"Block {block_no}, Line {line_no}, Word {word_no}:")
    print(f"Position: ({x0}, {y0}) to ({x1}, {y1})")
    print(f"Text: {word}")
    print("-----")
    print(len(words))

```
